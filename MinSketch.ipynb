```python
#let's assume that we have a list of strings A to E, I will use English ABC

strings = list("ABCDE")
numwid = 10
numhush = 15
stream = 500 
randomhash = np.random.choice(strings,stream)
```

```python
sketch = pd.DataFrame(np.zeros((numwid, numhush)))
```

```python
sketch.head(10)
```

```python
print(randomhash)
```

```python
for x in randomhash:
    for y in range(numwid):
        norm = mmh3.hash(x,sketch[y]) % numhush
        sketch[y,norm] += 1 
    
print(sketch)
```
